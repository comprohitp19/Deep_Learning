### Binary Classification - Booking Cancellation

### Step 1: Read the train dataset

```python
import pandas as pd
url1 = "https://raw.githubusercontent.com/utkarshg1/Neural-Net-Datasets/main/booking_train.csv"
df = pd.read_csv(url1)
df.head()
```

### Booking status is Target Feature
1. 0 - Not Cancelled
2. 1 - Cancelled

### Step 2 - Perform Basic Data Quality Checks

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

### Step 3 - Seperate X and Y(booking_status)

```python
X = df.drop(columns=['id', 'booking_status'])
Y = df[['booking_status']]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

```python
 Y.value_counts().plot(kind='bar')
```

### Step 4 - Apply Preprocessing on X

```python
 from sklearn.pipeline import Pipeline
 from sklearn.impute import SimpleImputer
 from sklearn.preprocessing import StandardScaler
```

```python
 num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='median')),
                          ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

### Step 5 : Perform Train Test Split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
 ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Step 6: Build the neural network

```python
xtrain.shape[1]
```

```python
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout
```

```python
# Initialize the model
model = Sequential()
# Add Input Layer 1
model.add(Input(shape=(xtrain.shape[1],)))
# Add Hidden Layer 1
model.add(Dense(units=64, activation='relu'))
# Add Dropout Layer1
model.add(Dropout(rate=0.4))
# Add Hidden Layer 2
model.add(Dense(units=32, activation='relu'))
# Add Dropout Layer 2
model.add(Dropout(rate=0.3))
# Add Output Layer
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
hist = model. fit(xtrain, ytrain, validation_split=0.2, epochs=60)
```

### Plot Learning Curve

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(hist.history['loss'], label='Train Loss')
plt.plot(hist.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Binary Crossentrophy')  # Corrected the typo here
plt.title('Learning Curve')
plt.legend()
plt.grid()
plt.show()
```

### Evaluate model in Train and Test

```python
model.evaluate(xtrain,ytrain)
```

```python
model.evaluate(xtest,ytest)
```

### Predict the results for train and test

```python
yprob_train = model.predict(xtrain)
yprob_test = model.predict(xtest)
```

### Plot the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cf = confusion_matrix(ytest,ypred_test)
cfd = ConfusionMatrixDisplay(cf)
cfd.plot()
```
