Regression with Keras

```python
import pandas as pd
url = "https://raw.githubusercontent.com/utkarshg1/Neural-Net-Datasets/main/PowerPlant.csv"
df = pd.read_csv(url)
df.head()
```

### Target feature here is PE (Power Plant Energy output)

###Step 2 - Perform Basic Quality Checks

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.duplicated().sum()/len(df)
```

### Step 3 - Seperate X and Y(PE)

```python
X = df.drop(columns=['PE'])
Y = df[['PE']]
```

```python
X.head()
```

### Step 4: Create a Preprocessing Pipeline for X

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer (strategy='mean' )),
                            ('scaler', StandardScaler ())]).set_output (transform='pandas')
```

```python
X_pre = num_pipe.fit_transform (X)
X_pre.head()
```

### Step 5 : Apply Train Test Split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
ytest.head()
```

### Step 5: Build the Keras neural network

```python
xtrain.shape[1]
```

```python
from keras.models import Sequential
from keras.layers import Dense, Input
```

```python
# Intialialize the sequential model
model = Sequential()

# Add the input layer
model.add (Input(shape=(xtrain.shape[1],)))

# Add the Hidden Layer 1
model.add(Dense(units=8, activation='relu'))

# Add the Hidden Layer 2
model.add(Dense(units=4, activation='relu'))

# Add the output Layer
model.add(Dense(units=1, activation='linear'))
```

```python
model.summary()
```

```python
model.compile(loss='mse', optimizer='adam' , metrics=['mae'])
```

```python
hist = model.fit(xtrain, ytrain, validation_split=0.2, epochs=100)
```

### Evaluate the model in train and test

```python
 model.evaluate(xtrain, ytrain)
```

```python
model.evaluate(xtest, ytest)
```
